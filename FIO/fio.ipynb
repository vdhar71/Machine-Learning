```python
import matplotlib.pyplot as plt
import numpy as np


cols = (2,48)

f1 = np.genfromtxt('fio1k.csv',delimiter=';',usecols=cols,dtype='str')
f2 = np.genfromtxt('fio4k.csv',delimiter=';',usecols=cols,dtype='str')

sync = []
sync1 = []
sync2 = []
asyn = []
asyn1 = []
asyn2 = []

for x in f1:
    if 'sync' in x[0]:
        s1 = float(x[1])
        sync1.append(s1)
    if 'async' in x[0]:
        a1 = float(x[1])
        asyn1.append(a1)

for x in f2:
    if 'sync' in x[0]:
        s2 = float(x[1])
        sync2.append(s2)
    if 'async' in x[0]:
        a2 = float(x[1])
        asyn2.append(a2)

fig = plt.figure()

ax1 = fig.add_subplot(111)

ax1.set_title("Write Throughput")    

x = np.linspace(-100,100,10000)
y1 = np.array(sync1)
y2 = np.array(asyn1)
y3 = np.array(sync2)
y4 = np.array(asyn2)

plt.plot(y1,c='b',label='Sync 1k IO')
plt.plot(y2,c='m',label='Async 1k IO')
plt.plot(y3,c='c',label='Sync 4k IO')
plt.plot(y4,c='r',label='Async 4k IO')

plt.show()

sync += sync1 + sync2
asyn += asyn1 + asyn2
ysync = np.array(sync)
yasync = np.array(asyn)
plt.plot(ysync,c='y',label='All Sync')
plt.plot(yasync,c='c',label='All Async')

plt.show()
plt.close('all')
```
