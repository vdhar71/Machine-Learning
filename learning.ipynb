```python

class line(object):
    
    def __init__(self,c1,c2):
        self.c1 = c1
        self.c2 = c2
        
    def distance(self):
        x1,y1 = self.c1
        x2,y2 = self.c2
        # Distance = sqrt((x2-x1)^2 + (y2-y1)^2)
        return ((x2-x1)**2 + (y2-y1)**2)**0.5
    
    def slope(self):
        x1,y1 = self.c1
        x2,y2 = self.c2
        # Slope = (y2-y1)/(x2-x1)
        return float((y2-y1)/(x2-x1))


cord1 = (2,2)
cord2 = (9,10)
line = line(cord1,cord2)

print("Distance: %4.5f Slope: %4.5f"% (line.distance(), line.slope()))
    
```

```python
class Circle:
    
    # Class object attribute
    pi = float(22/7)
    
    def __init__(self,radius=1):
        self.radius = radius
        self.perimeter = Circle.pi * 2 * self.radius #  Perimeter as Class object attribute
        
    def pm(self): # Perimeter as a Class method
        return Circle.pi * 2 * self.radius
    
    def area(self):
        r = self.radius
        return Circle.pi * r ** 2
    
    def set_radius(self,new_radius):
        """ This method takes radius parameter and resets the current radius to the new value"""
        self.radius = new_radius
        
    def get_radius(self):
        """ This method returns the current radius"""
        return self.radius
        
    
c = Circle(10)
for i in range(1,11):
    c.set_radius(i)
    print(c.get_radius(),'\t|%4.5f' % c.pm(), '\t|%4.5f' % c.area(),'\t|%4.3f'% c.perimeter)
    

```

```python
class Line:
    
    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return (((x2-x1)**2) + ((y2- y1)**2))** 0.5
    
    def slope(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return ((y2 - y1)/(x2 - x1))
    
        
c1 = (3,2)
c2 = (8,10)
li = Line(c1,c2)
print('Distance: %4.5f units'% li.distance(),'\t|Slope: %4.5f' % li.slope())
```

```python
class Cylinder:
    
    pi = 22.0/7
    
    def __init__(self,height=1,radius=1):
        self.h = height
        self.r = radius
        
    def volume(self):
        h = self.h
        r = self.r
        return Cylinder.pi * r**2 * h
    
    def sa(self):
        h = self.h
        r = self.r
        return Cylinder.pi * 2 * r * (r + h)
    
h,r = input().split(' ')
c = Cylinder(float(h),float(r))
print('Volume: %4.5f cubic units' % c.volume(), '|\tSurface Area: %4.5f square units' % c.sa())
```

```python
class Sphere:
    
    pi = 22.0/7
    
    def __init__(self,radius):
        self.r = radius
        
    def volume(self):
        r = self.r
        return Sphere.pi * 4.0/3 * r**3
    
    def sa(self):
        r = self.r
        return Sphere.pi * 4 * r**2
    

sp = Sphere(float(input('Enter the radius:')))
print('Volume: %4.5f' % sp.volume(), '|\tSurface Area: %4.5f' % sp.sa())
    
```

```python
class Ellipse:
    
    def __init__(self,xa,yb):
        self.x = xa
        self.y = yb
        
    def felip(self):
        x,a = self.x
        y,b = self.y
        return ((y/b)**2 + (x/a)**2) == 1
    
ax = (1,3)
by = (4,4)
eo = Ellipse(ax,by)
type(eo)
if eo.felip():
    print('It is an ellipse')
else:
    print('Not an ellipse')
        
```

```python
class Hyperbola:
    
    def __init__(self,xa,yb):
        self.x = xa
        self.y = yb
        
    def hyp(self):
        x,a = self.x
        y,b = self.y
        return ((y/b)**2 + (x/a)**2) > 1
    
ax = (5,4)
by = (4,5)
eo = Hyperbola(ax,by)
type(eo)
if eo.hyp():
    print('It is a hyperbola')
else:
    print('Not a hyperbola')
```

```python
def ask4int():
    
    while True:
        try:
            val = int(input('Enter an integer:'))
            num = int(input('Enter the divisor:'))
            val /= num

        except ValueError as err:
            print('That is not an integer, please try again! {0}'.format(err))
            continue
            
        except ZeroDivisionError as err:
            print('Divide by zero {0} {1}'.format(val,err))
            continue
            
        else:
            print('Integer entered')
            break
            
        finally:
            print('This block is always executed')
            
        print(val)
        
```

```python
def check_power_n(dest,power,start):
    
    if dest == 0:
        return 1
    count = 0
    plist = [pow(t,power) for t in range(1,dest) if pow(t,power) <= dest]
    destlen = len(plist)+1
    for i in range(start+1,destlen):
        if pow(i,power) <= dest:
            count += check_power_n(dest-pow(i,power),power,i)
   
    return count
                
while True:
    try:
        x=int(input('Enter x:'))
    except ValueError as e:
        print('Invalid input {0}'.format(e))
    if x < 1 or x > 1000:
        print('Invalid input, please try again')
        continue
    else:
        print('x:', x)
        break
        
while True:
    try:
        n=int(input('Enter n:'))
    except ValueError as e:
        print('Invalid input {0}'.format(e))
    if n < 2 or n > 10:
        print('Invalid input, please try again')
        continue
    else:
        print('n:',n)
        break

powertest = check_power_n(x,n,0)
print(powertest)
```

```python
""" Bubble sort """
def bubblesort(a,n):
    numswap = 0
    for i in range(n):
        for j in range(n-1):
            if a[j] > a[j+1]:
                a[j],a[j+1] = a[j+1],a[j]
                numswap += 1
                
    return numswap
                
#Main
n = int(input().strip())
a = list(map(int, input().strip().split(' ')))
swap = bubblesort(a,n)
            
print('Array is sorted in',numswap,'swaps.',a)
print('First Element:',a[0])
print('Last Element:',a[-1])
```

```python
def isfib(n):
    flag = 0
    a,b = 1,1
    while a<n:
        a,b = b, a+b
    if a == n:
        flag = 1
    return flag

# Main
t = int(input().strip())
li = []
for i in range(t):
    num = int(input().strip())
    li.append(num)
    
for x in li:
    fib = isfib(x)
    if fib:
        print(x,'IsFibo')
    else:
        print(x,'IsNotFibo')
```
